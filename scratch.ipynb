```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python

def generateTrajectories(r, phi, d_phi=0.05, n_rays=80):
    x_traj = []
    y_traj = []
    half_ray = n_rays // 2

    for i in range(2*half_ray):
        fac = 1 if i < half_ray else -1
        
        u = 1/r
        u_dot = u*np.tan(np.pi*(0.4 + 0.1*i/(half_ray-1))) #d_u/d_phi

        x = [np.cos(phi)/u]
        y = [np.sin(phi)/u]

        for i in range(1000):
            u_dot += (3*u**2 - u)*d_phi*fac
            u += u_dot*d_phi*fac
            phi += d_phi*fac

            if u < 0: break
            if u > 0.5: break

            x.append(np.cos(phi)/u)
            y.append(np.sin(phi)/u)

        x_traj.append(x)
        y_traj.append(y)
    
    return (x_traj, y_traj)
```

```python
(x_traj, y_traj) = generateTrajectories(40, 2.3)
fig, ax = plt.subplots(figsize=(8, 8))

for i in range(2*n_rays):
    c = "#555555" if x_traj[i][-1]**2 + y_traj[i][-1]**2 < 4.5 else "#ff80ff"
    ax.plot(x_traj[i], y_traj[i], color=c)

ax.set_xbound(-7, 7)
ax.set_ybound(-7, 7)
```
