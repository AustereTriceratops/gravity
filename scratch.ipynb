```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
d_phi = 0.05
n_rays = 50

x_traj = []
y_traj = []

for i in range(2*n_rays):
    fac = 1 if i < n_rays else -1
    r = 4
    u = 1/r

    phi = 0
    u_dot = -np.tan(0.0001 + np.pi*(i/(n_rays-1) - 1/2))*u #d_u/d_phi

    x = [np.cos(phi)/u]
    y = [np.sin(phi)/u]

    for i in range(1000):
        u_dot += (3*u**2 - u)*d_phi*fac
        u += u_dot*d_phi*fac
        phi += d_phi*fac

        if u < 0: break
        if u > 0.5: break

        x.append(np.cos(phi)/u)
        y.append(np.sin(phi)/u)

    x_traj.append(x)
    y_traj.append(y)
```

```python
d_phi = 0.02
n_rays = 50

x_traj = []
y_traj = []

for i in range(2*n_rays):
    fac = 1 if i < n_rays else -1
    r = 40
    u = 1/r

    phi = 2.3
    u_dot = u*np.tan(np.pi*(0.4 + 0.1*i/(n_rays-1))) #d_u/d_phi

    x = [np.cos(phi)/u]
    y = [np.sin(phi)/u]

    for i in range(1000):
        u_dot += (3*u**2 - u)*d_phi*fac
        u += u_dot*d_phi*fac
        phi += d_phi*fac

        if u < 0: break
        if u > 0.5: break

        x.append(np.cos(phi)/u)
        y.append(np.sin(phi)/u)

    x_traj.append(x)
    y_traj.append(y)
```

```python
fig, ax = plt.subplots(figsize=(8, 8))

for i in range(2*n_rays):
    c = "#555555" if x_traj[i][-1]**2 + y_traj[i][-1]**2 < 4.5 else "#ff80ff"
    ax.plot(x_traj[i], y_traj[i], color=c)

ax.set_xbound(-7, 7)
ax.set_ybound(-7, 7)
```
